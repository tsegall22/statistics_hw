## Confidence Interval

```python
# This code is taken from "StackOverFlow" : 
# https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data
def mean_confidence_interval(data_vector, confidence=0.95):
    a = 1.0 * np.array(data_vector)
    n = len(a)
    m, se = np.mean(a), st.sem(a)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return m-h, m+h
```

```python
for feature in numeric_features:
    print('Feature: {}'.format(feature))
    
    l1_start, l1_end = mean_confidence_interval(df_sick[feature])
    l2_start, l2_end = mean_confidence_interval(df_healthy[feature])
    
    if (l1_start > l2_end):
        print("Sick group mean is larger with confidence of 95%".format(feature))
    elif (l2_start > l1_end):
        print("Healthy group mean is larger with confidence of 95%".format(feature))
    else:
        print("CI of means overlap with confidence 95%".format(feature))
        
    data = [df_sick[feature], df_healthy[feature]]
    labels = ['Liver Patient', 'Non Liver Patient']
        
    fig, axs = plt.subplots()
    sns.barplot(x='Label', y=feature, data=df)
    plt.show()    
    fig.savefig('2_ConfidenceInterval_{}'.format(feature.replace('/', '')))
```

## Histograms

```python
def draw_hist_and_mle(feature, distribution, bins=10):
    data_vector = df[feature]
    
    # fit mle
    fit_params = distribution.fit(data_vector)
    mle = distribution(*fit_params)
    
    # plot mle
    x = np.linspace(data_vector.min(), data_vector.max(), int((data_vector.max() - data_vector.min()) * 10))
    y = mle.pdf(x)
    
    fig, axs = plt.subplots()
    
    data_vector.hist(density=True, bins=bins, label='Histogram')
    plt.plot(x, y, label='MLE')
    plt.title('{} Histogram and MLE'.format(feature))
    plt.legend(loc='best')
    plt.show()
    fig.savefig('2_Histogram_MLE_{}'.format(feature.replace('/', '')))
    
def draw_hist_with_normal_mle(feature, bins=50):
    return draw_hist_and_mle(feature, st.norm, bins)
    
def draw_hist_with_log_normal_mle(feature, bins=50):
    return draw_hist_and_mle(feature, st.lognorm, bins)

def draw_hist_with_expon_mle(feature, bins=50):
    return draw_hist_and_mle(feature, st.expon, bins)
```

```python
draw_hist_with_log_normal_mle('TB')
draw_hist_with_expon_mle('DB')
draw_hist_with_log_normal_mle('Alkphos')
draw_hist_with_expon_mle('Sgpt')
draw_hist_with_expon_mle('Sgot')
draw_hist_with_normal_mle('TP')
draw_hist_with_normal_mle('ALB')
draw_hist_with_normal_mle('A/G Ratio')
```

## Correlations

To find correlation, we will use the stats.pearsonr.
stats.pearsonr returns the correlation r based on 2 data vectors, and the P-Value 
under the null model that there is no correlation between the variables. 

```python
def are_correlated(data_table, f1, f2):
    r, p = st.pearsonr(data_table[f1], data_table[f2])
    return (p < 0.05), p, r
    
def find_significant_correlations(data_table, feature_list):
    correlations = []
    
    numOfComb = 0
    for f1, f2 in itertools.combinations(feature_list, 2):
        numOfComb+=1
        if f1 == f2:
            continue

        correlated, p, r = are_correlated(data_table, f1, f2)
        if correlated:
            correlations.append((f1, f2, p, r))
    
    #Avoid Multipe Testing issue by useing Bonfferroni correction
    print ("Number of tests: %u" %numOfComb)
    print ("Before consider multiple-testing, number of correlations found: %u" %len(correlations))
    i=0
    while (i<len(correlations)):
        if (correlations[i][2]*numOfComb > 0.05):
            del correlations [i]
        else:
            i+=1
            
    print ("After consider multiple-testing, number of correlations found: %u" %len(correlations))
    return correlations
```

```python
for feature1, feature2, p, r in find_significant_correlations(df, numeric_features):
    print("Correlation found between {} and {}. p-value={}, ro={}".format(feature1, feature2, format(p, '.2'), format(r, '.2')))
```

Are there pairs that are significantly correlated in males but not in females? The opposite? How about healthy vs disease?

# Question 5

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
T = np.array(
    [
        [0.5, 0.25, 0, 0, 0, 0.25],
        [0.25, 0.5, 0.25, 0, 0, 0],
        [0, 0.25, 0.5, 0.25, 0, 0],
        [0, 0, 0.25, 0.5, 0.25, 0],
        [0, 0, 0, 0.25, 0.5, 0.25],
        [0.25, 0, 0, 0, 0.25, 0.5]
    ])
```

```python
DICE_RESULT = np.arange(1, 7)
```

```python
def markov_clt(length):
    # First roll is a uniform distribution 
    p = np.array([1.0 / len(DICE_RESULT)] * len(DICE_RESULT)) 
    
    trajectory = []
    
    for i in range(length):
        dice_result = np.random.choice(DICE_RESULT, replace=False, p=p)
        trajectory.append(dice_result)
        p = T[dice_result - 1]
    
    return trajectory
```

## Problem A

```python
num_trajectories = 1000
length = 20

trajectories = []

for i in range(num_trajectories):
    trajectories.append(markov_clt(length))
    
trajectories = np.array(trajectories)

trajectories_avg = np.average(trajectories, axis=1)

mean = np.mean(trajectories_avg)
std = np.std(trajectories_avg)

fig, axs = plt.subplots()
axs.hist(trajectories_avg, bins=100, density=True, label='histogram')

x = np.linspace(np.min(trajectories_avg), np.max(trajectories_avg), 100)
y = norm.pdf(x, mean, std)
axs.plot(x, y, label='Normal Fit')

axs.set_title('Trajectories average histogram')

plt.legend(loc='best')
plt.show()
fig.savefig('5A_ii.png')

print('Mean: %f, Std: %f' % (mean, std))
```
